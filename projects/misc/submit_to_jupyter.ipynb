# Import Required Libraries
Import libraries needed for interacting with Jupyter, such as jupyter_client.

```python
from jupyter_client import KernelManager
import queue
```

# Connect to Jupyter Interactive Session
Establish a connection to a running Jupyter session and verify the kernel is available.

```python
# Start a kernel
km = KernelManager()
km.start_kernel()
kc = km.client()
kc.start_channels()
# Wait for the kernel to be ready
kc.wait_for_ready()
```

# Create and Submit Code Cells
Programmatically define code cells and submit them to the interactive session for execution.

```python
# Define the code to submit
code = """import pandas as pd
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt"""

# Execute the code
msg_id = kc.execute(code)
```

# Capture Output from Interactive Session
Read the execution results and output from the submitted cells.

```python
# Capture the output
reply = kc.get_shell_msg(timeout=10)
print("Execution reply:", reply)

# Get stdout
stdout_msg = kc.get_iopub_msg(timeout=10)
print("Output:", stdout_msg)
```

# Handle Execution Errors
Detect and report errors returned by the interactive session during code execution.

```python
# Check for errors
if reply['content']['status'] == 'error':
    print("Error:", reply['content']['traceback'])
else:
    print("Execution successful")

# Shutdown the kernel
kc.stop_channels()
km.shutdown_kernel()
```
